### Filtering Using String Methods

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\hunte\Documents\Analyst Workshop\Pandas For Data Analysis\3. Filtering and Ordering\laptop_prices.csv",
                encoding = 'ISO-8859-1')

pd.set_option('display.max.rows', 15)
pd.set_option('display.max.columns', 15)
```

```python
df[df['Product'].str.contains('MacBook')]
```

```python
df[~df['Product'].str.contains('MacBook')]
```

```python
df[df['Product'].str.contains('MacBook') == True]
```

```python
df[df['Product'].str.contains('MacBook') == False]
```

```python
df[(df['Product'].str.contains('MacBook')) & (df['Memory'].str.contains('SSD'))] 
```

```python
df[df['Product'].str.startswith('Mac')]
```

```python
# Remember this function cant work with integers. Just values with strings. It can be strings mixed with numbers

df[df['Product'].str.endswith('4')]
```

```python
df[df['Price_usd'].duplicated()]
```

```python
df[df['Product'].str.match('Mac')]
```

```python
df[df['Product'].str.match('Mac.*Air')]
```
