### Ordering in DataFrames

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\hunte\Documents\Analyst Workshop\Pandas For Data Analysis\3. Filtering and Ordering\laptop_prices.csv",
                encoding = 'ISO-8859-1')

pd.set_option('display.max.rows', 15)
pd.set_option('display.max.columns', 15)
```

```python
df
```

```python
df.sort_values(by = 'Price_usd')
```

```python
df.sort_values(by = 'Price_usd', ascending = True)
```

```python
df.sort_values(by = 'Price_usd', ascending = False)
```

```python
df[df['Product'] == 'Blade Pro'].sort_values(by = 'Price_usd', ascending = False)
```

```python
df[df['Company'] == 'Razer'].sort_values(by = ['Inches','Price_usd'], ascending = [True,False])
```
