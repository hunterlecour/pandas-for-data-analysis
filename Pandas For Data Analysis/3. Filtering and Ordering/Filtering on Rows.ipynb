### Filtering on Rows

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\hunte\Documents\Analyst Workshop\Pandas For Data Analysis\3. Filtering and Ordering\laptop_prices.csv",
                encoding = 'ISO-8859-1')

pd.set_option('display.max.rows', 15)
pd.set_option('display.max.columns', 15)

df
```

```python
# This attempt at filtering throws a KeyError because df['Company'] returns the entire column instead of a filter condition.

# Explanation: The correct usage is to apply a boolean condition, like checking where the company name equals a specific value.


df[df['Company']]
```

```python
# This filter extracts all rows where the Company column contains 'Apple'.

# How it works:
# df['Company'] == 'Apple' creates a boolean series where True indicates rows with 'Apple'.
# The DataFrame returns only those rows where the condition is True.

df[df['Company']== 'Apple']
```

```python
# Filtering with Multiple Conditions Using & (AND)

# This filter selects rows where both conditions are true:
# The company is 'Apple'.
# The RAM is '16GB'.
# & Operator: Ensures that both conditions must be met for a row to be included.


df[(df['Company']== 'Apple') & (df['Ram'] == '16GB')]
```

```python
# Filtering with | (OR) Operator

# This filter selects rows where either of the following conditions is true:
# Company is 'Apple'.
# RAM is '16GB'.
# | Operator: Ensures that either condition being true will include a row.


df[(df['Company']== 'Apple') | (df['Ram'] == '16GB')]
```

```python
df[(df['Company']== 'Apple') | (df['Company'] == 'Dell')]
```

```python
# Using .isin() for Multiple Values

# isin(): Checks whether the Company column contains either 'Apple' or 'Dell'.
# This method is more efficient than using multiple | operators for long lists of values.

df[df['Company'].isin(['Apple','Dell'])]
```

```python
# Filtering with Comparison Operators

# This filter selects laptops where the price is 1500 USD or higher.
# Comparison Operators:
# >=: Greater than or equal to.
# Similar operators: <, <=, ==, !=.

df[df['Price_usd']>= 1500]
```

```python
df.loc[df['Company']=='Apple']
```

```python
# df.loc[:5]:

# Retrieves rows by label index up to (and including) the label 5.
# This means it will select the first six rows (assuming the DataFrame has label-based indexing starting from 0).
# Uses labels, not numerical positions.

df.loc[:5]
```

```python
# df.iloc[:5]:

# Retrieves the first 5 rows using their integer positions (i.e., rows at positions 0 to 4).
# Uses integer-based indexing


df.iloc[:5]
```

```python
# df.iloc[[0, 5, 10, 20]]:

# Retrieves rows at positions 0, 5, 10, and 20 by integer location.


df.iloc[[0,5,10,20]]
```

```python
# df.filter(items=[0, 5, 10, 20], axis=0):

# Tries to filter the DataFrame using labels, not positions.
# It expects the index labels of the DataFrame to match the list [0, 5, 10, 20].

df.filter(items = [0,5,10,20], axis = 0)
```

```python
# why wont this work correctly

# Python thinks you're asking for columns named 0, 5, 10, and 20. But your DataFrame doesn't have columns with those names. So, it returns an empty DataFrame, showing only the index numbers without any data.
# If you want to filter rows by their index numbers, not columns add axis=0


df.filter(items = [0,5,10,20])
```
