```python
# Querying the Data using df.query() Method:
# The query() method allows filtering the DataFrame based on conditions, similar to SQL's WHERE clause.
# Purpose: Displays rows where the Price_usd is greater than 1200.
# How it works: The query method evaluates the expression and returns only matching rows.

# its pulling all the columns where price_usd > 1200

df.query("Price_usd > 1200")
```

```python
# or operator returns rows where Company is either 'Apple' or 'Dell'

df.query("Company == 'Apple' or Company == 'Dell'")
```

```python
# Explanation: This condition selects laptops that are either:
# Made by Apple, or
# Have a price greater than 1500 USD.

df.query("Company == 'Apple' or Price_usd > 1500")
```

```python
# and operator ensures that both conditions are satisfied.

df.query("Company == 'Apple' and Price_usd > 1500")
```

```python
# SQL Queries on DataFrame Using pandasql:

# sqldf() allows SQL queries on the DataFrame.
# Explanation: This SQL query selects all rows where the Company is 'Apple'.
# Benefit: For users familiar with SQL, this provides an intuitive way to manipulate DataFrames.

psql.sqldf("SELECT * FROM df WHERE Company = 'Apple'")
```

```python
psql.sqldf("SELECT * FROM df WHERE Company = 'Apple' And Inches > 14")
```
